```python
import pandas as pd
import numpy as np
```

```python
df_hr=pd.read_csv('/content/Dataset - HR Employee Attrition.csv')
df_hr.head()
```

```python
df_hr['EmployeeCount'].unique()
```

```python
df_hr['StandardHours'].unique()
```

```python
# prompt: drop columns employeecount and standardhours

df_hr.drop(columns=['EmployeeCount', 'StandardHours'], inplace=True)
df_hr.head()
```

```python
# so we can drop these two columns as there are no unique values and they cause NAN after normalisation
```

```python
# so 2 classes yes and no
df_hr['Attrition'].value_counts()
```

```python
# prompt: search column with attrition
df=df_hr.iloc[:,1:35]
df.head()
```

```python
df.describe()
```

```python
#using min max scaler
df.info()
```

```python
df_n=df.iloc[:,1:35]   #all the input data
df_n.head()
```

```python
df_n = pd.get_dummies(df_n, drop_first=True).astype(int)
df_n.head()
```

```python
df_n.info()
```

```python
def norm_func(i):
    x = (i-i.min())	/ (i.max()-i.min())
    return (x)
```

```python
df_n = norm_func(df_n.iloc[:, 0:])
df_n.describe()
```

```python
#  find rows with missing values in df_n

df_n.isnull().sum()
```

```python
# divide datat types into x and y
X = np.array(df_n.iloc[:,:]) # Predictors
Y = np.array(df_hr['Attrition']) # Target
```

```python
X.shape
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#NOTE ININTIALLY IT WAS GIVING PROBLEM HERE CUZ AFTER NORMALISATION EMPLOYEECOUNT AND STANDARD HOURS HAD NAN VALUE. SINCE THEY HAD
# ALL THE VALUES SAME ONLY FOR ALL ROWS I DROPPED THOSE TWO COLUMNS

knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, Y_train)

pred = knn.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, pred))
pd.crosstab(Y_test, pred, rownames = ['Actual'], colnames= ['Predictions'])
```

```python
# error on train data
pred_train = knn.predict(X_train)
print(accuracy_score(Y_train, pred_train))
pd.crosstab(Y_train, pred_train, rownames=['Actual'], colnames = ['Predictions'])
```

```python
# creating empty list variable
acc = []

# running KNN algorithm for 3 to 50 nearest neighbours(odd numbers) and
# storing the accuracy values

for i in range(3,50,2):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train, Y_train)
    train_acc = np.mean(neigh.predict(X_train) == Y_train)
    test_acc = np.mean(neigh.predict(X_test) == Y_test)
    Difference=train_acc-test_acc
    acc.append([i,train_acc, test_acc,Difference])
```

```python
df_new=pd.DataFrame(acc)
```

```python
df_new.columns='K','train_acc','test_acc','difference'
df_new
```

```python
import matplotlib.pyplot as plt # library to do visualizations

# train accuracy plot
plt.plot(np.arange(3,50,2),[i[1] for i in acc],"ro-")

# test accuracy plot
plt.plot(np.arange(3,50,2),[i[2] for i in acc],"bo-")
```

```python
# Thus k=5,9,13 gives closest values
```

```python
knn = KNeighborsClassifier(n_neighbors = 9)
knn.fit(X_train, Y_train)

pred = knn.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, pred))
pd.crosstab(Y_test, pred, rownames = ['Actual'], colnames= ['Predictions'])
```
