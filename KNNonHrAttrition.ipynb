```python
import pandas as pd

# Convert the NumPy array back to a Pandas DataFrame
df_n = pd.DataFrame(df_n) 

# Now you can use isnull()
df_n.isnull().sum()
```

```python
# divide datat types into x and y
X = np.array(df_n.iloc[:,:]) # Predictors
Y = np.array(df_hr['Attrition']) # Target
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming X_train is your data that you want to scale
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_test=scaler.fit_transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_test.shape
```

```python
#NOTE ININTIALLY IT WAS GIVING PROBLEM HERE CUZ AFTER NORMALISATION EMPLOYEECOUNT AND STANDARD HOURS HAD NAN VALUE. SINCE THEY HAD
# ALL THE VALUES SAME ONLY FOR ALL ROWS I DROPPED THOSE TWO COLUMNS

knn = KNeighborsClassifier(n_neighbors = 9)
knn.fit(X_train, Y_train)

pred = knn.predict(X_test)
pred.shape
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, pred))
pd.crosstab(Y_test, pred, rownames = ['Actual'], colnames= ['Predictions'])
```

```python
# error on train data
pred_train = knn.predict(X_train)
print(accuracy_score(Y_train, pred_train))
pd.crosstab(Y_train, pred_train, rownames=['Actual'], colnames = ['Predictions'])
```
